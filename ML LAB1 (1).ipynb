```python
# ques 1.
def vowels(s):
    a=0
    b=0
    for i in s:
        if i in ['a','e','i','o','u']:
             a=a+1
        else:
            b=b+1
    return a,b
s=input("enter a string")
a,b= (vowels(s))
print("vowels: ",a)
print("consonants: ",b)




#ques 2.
def read_matrix(name):
    print(f"Enter rows of matrix {name}. Enter each row as space-separated numbers.")
    print("Type 'done' when you are finished.")
   
    matrix = []
    while True:
        row = input("Enter row: ")
        if row.lower() == 'done':
            break
        matrix.append([int(num) for num in row.split()])
    return matrix


def multiply_matrices(A, B):
    # Get dimensions
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])
   
   
    if cols_A != rows_B:
        return None  

    
    result = [[0] * cols_B for _ in range(rows_A)]

    
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]
   
    return result

def print_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))


print("Matrix A:")
matrix_a = read_matrix('A')

print("Matrix B:")
matrix_b = read_matrix('B')


product = multiply_matrices(matrix_a, matrix_b)


if product is None:
    print("Error: Matrices cannot be multiplied.")
else:
    print("Product of matrix A and B:")
    print_matrix(product)


    
    
# ques 3.
def commonEle():
    L1 = []
    L2 = []
    s1 = int(input("Enter size of list L1"))
    s2 = int(input("Enter size of list l2"))

    for i in range(s1):
        a=int(input())
        L1.append(a)
    
    for i in range(s2):
        a=int(input())
        L2.append(a)
    L3=[]
    for i in L1:
        if i in L2:
            L3.append(i)
            
    return L3

print("The common elements are: ",commonEle())






#ques 
def matrix():
    
    row = int(input("Enter number of rows: "))
    col = int(input("Enter number of coloumns: "))

    m=[]
    for i in range(row):
        a=[]
        for j in range(col):
            b=int(input())
            a.append(b)
        m.append(a)
    print(m)

    temp = [[m[i][j] for i in range(row)] for j in range(col)]
    for i in range(row):
        for j in range(col):
            if i!=j:
                temp[i][j]=m[j][i]
    
    return temp
print(matrix())
```
